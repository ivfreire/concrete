```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib.animation import FuncAnimation
```

```python
plt.rcParams['figure.figsize'] = (10, 10)
```

# Visualization of particles systems

## Simple Collision

```python
data = np.loadtxt('../data/collision.txt')
```

```python
ax = plt.axes(projection='3d')

ax.set(xlim=(0, 1), ylim=(0, 1), zlim=(0, 1), xlabel='x', ylabel='y', zlabel='z')

for i in range(2):
    n = 140
    x, y, z = data[n:n+10].transpose()[3*i:3*(i+1)]
    ax.plot3D(x, y, z)

plt.show()
```

Animation

```python
fig = plt.figure(figsize=(5,5))
ax = plt.axes(projection='3d')

ax.set(xlim=(0, 1), ylim=(0, 1), zlim=(0, 1), xlabel='x', ylabel='y', zlabel='z')


def update(frame):
	lines = []
	for i in range(2):
		n = frame
		x, y, z = data[n:n+1].transpose()[3*i:3*(i+1)]
		plot = ax.plot3D(x, y, z)
		lines.append(plot)
	return lines

ani = FuncAnimation(fig=fig, func=update, frames=200, interval=10)
ani.save('collision.gif', writer='imagemagick')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['figure.figsize'] = (12,8)
sns.set_theme()
```

```python
def hertz_mindlin(K, R, d): return (4/3) * K * R * d**(3/2)
```

```python
x = np.linspace(0, 1)
```

```python
plt.plot(x, hertz_mindlin( 1E5, 1, x ))
plt.show()
```

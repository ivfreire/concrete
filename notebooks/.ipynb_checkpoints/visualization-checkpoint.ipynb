# Visualization of particles systems

## Simple Collision

```python
data = np.loadtxt('../data/collision.txt')
```

```python
ax = plt.axes(projection='3d')

ax.set(xlim=(0, 1), ylim=(0, 1), zlim=(0, 1), xlabel='x', ylabel='y', zlabel='z')

for i in range(2):
    n = 900
    x, y, z = data[n:n+20].transpose()[3*i:3*(i+1)]
    ax.plot3D(x, y, z)

plt.show()
```
